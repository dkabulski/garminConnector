```
 
rhrList = []
for jsonFile in outputlist:
  for day in jsonFile: #this returns dictionary
    rhrdict = {}
    #print(day)
    for entry in day.items():
      if isinstance(entry[1],dict):
        if entry[1].get('date') is not None:
          rhrdict[entry[0]] = entry[1].get('date')
          #for anything else (allDayStress & bodyBattery just return dict)
        else: 
          rhrdict[entry[0]] = "removed" #entry[1]
      else:
        rhrdict[entry[0]] = entry[1]
    rhrList.append(rhrdict)
```
